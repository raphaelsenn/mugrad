```python
import matplotlib.pyplot as plt

from sklearn.datasets import make_regression

from tensor import Tensor
from nn import Linear, Sequential
```

```python
X, y = make_regression(n_samples=1000, n_features=2, noise=20)
print(X.shape)
plt.scatter(X[:, 0], y)
```

```python
model = Linear(2, 1)
X = Tensor(X)
y = Tensor(y.reshape(-1, 1))
for epoch in range(10):
    model.zero_grad() 
    y_pred = model.forward(X)
    loss = ((y - y_pred)**2).mean()
    # print(X.data.shape, y.data.shape, y_pred.data.shape, loss.data.shape)
    loss.backward()
    for p in model.parameters():
        p.data = p.data - 0.4 * p.grad
        p.data = p.data - 0.4 * p.grad

    if epoch % 1 == 0: print(f'epoch: {epoch}\tloss: {loss.data.item()}')
```
